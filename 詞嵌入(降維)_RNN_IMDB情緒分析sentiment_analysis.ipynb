```
import glob
import pandas as pd
cname = ["article", "target"]
train_df = pd.DataFrame(columns=cname)
neg = glob.glob("aclImdb/train/neg/*.txt")
for p in neg:
  f = open(p, "r", encoding="utf-8")
  article = f.read()
  f.close()
  s = pd.Series([article, 0], index=cname) # 只有一維用Series([])
  train_df = train_df.append(s, ignore_index=True) # 要設定回去
pos = glob.glob("aclImdb/train/pos/*.txt")
for p in pos:
  f = open(p, "r", encoding="utf-8")
  article = f.read()
  f.close()
  s = pd.Series([article, 1], index=cname) # 只有一維用Series([])
  train_df = train_df.append(s, ignore_index=True) # 要設定回去
train_df
```

```
cname = ["article", "target"]
test_df = pd.DataFrame(columns=cname)
neg = glob.glob("aclImdb/test/neg/*.txt")
for p in neg:
  f = open(p, "r", encoding="utf-8")
  article = f.read()
  f.close()
  s = pd.Series([article, 0], index=cname) # 只有一維用Series([])
  test_df = test_df.append(s, ignore_index=True) # 要設定回去
pos = glob.glob("aclImdb/test/pos/*.txt")
for p in pos:
  f = open(p, "r", encoding="utf-8")
  article = f.read()
  f.close()
  s = pd.Series([article, 1], index=cname) # 只有一維用Series([])
  test_df = test_df.append(s, ignore_index=True) # 要設定回去
test_df
```

每一個詞都對每個語意有貢獻，所以是2000維度 * 32個維度 = 64000

輸出層如果只有一個函數，可以選用sigmoid函數

輸出層，如果有兩個以上輸出神經，請選用softmax函數

如果是中間層，只有一種選擇relu函數

```
from keras.models import Sequential
from keras.layers import Embedding
from keras.layers import Dense, Flatten, Dropout
model = Sequential()
model.add(Embedding(2000, 32, input_length=100)) # 降維，每一個詞維度2000,語意維度32或64,輸入100個字
model.add(Flatten()) # 攤開
model.add(Dense(256, activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(1, activation="sigmoid"))
model.summary()
```

```
from keras.preprocessing.text import Tokenizer # 找出最常用詞彙
tok = Tokenizer(num_words=2000) # 找出最常用2000字
tok.fit_on_texts(train_df["article"])
# print(tok.word_index)
```

```
train_df_seq = tok.texts_to_sequences(train_df["article"]) # 每篇文章抓出來，只留下常用兩千字
test_df_seq = tok.texts_to_sequences(test_df["article"])
pd.DataFrame(train_df_seq)
```

```
from keras.preprocessing.sequence import pad_sequences # 截長補短，因為要丟進mlp，長度要固定
train_df_pad = pad_sequences(train_df_seq, maxlen=100)
test_df_pad = pad_sequences(test_df_seq, maxlen=100)
pd.DataFrame(train_df_pad)
```

輸出一個神經元選擇:binary_crossentropy，多個神經元選擇categorical_crossentropy adam小彈珠優化法，無腦選

```
model.compile(loss="binary_crossentropy",
       optimizer="adam",
       metrics=["accuracy"])
```

```
model.fit(train_df_pad, train_df["target"],
     batch_size=200, epochs=3,
     validation_split=0.1, verbose=2) # 看100~200筆資料做一次梯度下降，完整訓練?次找到平滑的地方最棒，把10%當作考試材料
```

```
model.evaluate(test_df_pad, test_df["target"]) # 返回的是损失值和你选定的指标值（例如，精度accuracy）
```

```
from keras.models import Model
# model.layers[0].output拿出tensor
partial = Model(inputs=model.input, outputs=model.layers[0].output) # 拿出embedding_3層
partial.summary()
```

```
reverse_dict = {tok.word_index[k]:k for k in tok.word_index} # 轉換辭典
print("Embedding前:", reverse_dict[test_df_pad[0][-1]])
print("Embedding後:", partial.predict(test_df_pad)[0][-1])
```

```
# 加入SimpleRNN層
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN
from keras.layers import Dense, Flatten, Dropout
model = Sequential()
model.add(Embedding(2000, 32, input_length=100)) # 降維，每一個詞維度2000,語意維度32或64,輸入100個字
model.add(SimpleRNN(16)) # 循環層，使用tanh函數，輸出[-1, 1]之間(非機率可使用)
model.add(Flatten()) # 攤開
model.add(Dense(256, activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(1, activation="sigmoid"))
model.summary()
```

```
model.compile(loss="binary_crossentropy",
       optimizer="adam",
       metrics=["accuracy"])
```

```
model.fit(train_df_pad, train_df["target"],
     batch_size=200, epochs=3,
     validation_split=0.1, verbose=2) # 看100~200筆資料做一次梯度下降，完整訓練?次找到平滑的地方最棒，把10%當作考
```

```
model.evaluate(test_df_pad, test_df["target"]) # 返回的是损失值和你选定的指标值（例如，精度accuracy）
```
